## #1.1 Одномерный градиентный спуск

Реализуйте алгоритм одномерного градиентного спуска по шаблону, предложенному ниже. Можно редактировать шаблон на своё усмотрение. В одномерном пространстве градиент совпадает с производной функции.

```python
def gradient_descent(
    func, grad_func,
    initial_guess: float,
    learning_rate: float,
    epsilon: float = 1.0e-6,
    maxiter: int = 100000) -> tuple[float, float, bool]:
    """
    Params
    ------
    func : оптимизируемая функция с сигнатурой func(x:float)->float
    grad_func : градиент оптимизируемой функции с сигнатурой func(x:float)->float
    initial_guess : начальное приближение для аргумента оптимизируемой функции
    learning_rate : фиксированная скорость обучения
    epsilon : порог для пренебрежимо малой разности между значениями функции
    maxiter : максимальное число итераций

    Returns
    -------
    opt_arg : оптимальное значение аргумента
    opt_func : оптимальное значение функции
    flag : True если достигнута сходимость (выполнен критерий с epsilon) за maxiter итераций
    """

    arg = initial_guess
    prev_func = func(arg)

    for i in range(maxiter):
        direction = grad_func(arg)
        arg = arg - learning_rate * direction

        curr_func = func(arg)
        if abs(curr_func - prev_func) < epsilon:
            opt_arg, opt_func, flag = arg, curr_func, True
            return opt_arg, opt_func, flag

        prev_func = curr_func

    opt_arg, opt_func, flag = arg, func(arg), False
    return opt_arg, opt_func, flag
```
